```python
#Import Numpy & Pandas
import pandas as pd
import numpy as np
```

```python
#DataFrame
cricket_df = {"Virat":np.random.randint(50,100,10),
              "Root":np.random.randint(50,100,10),
              "Rohit":np.random.randint(50,100,10)
    
}

cricket_df

my_data = pd.DataFrame(cricket_df)
my_data
```

```python
#head & tail 
print(my_data.head(n=2))
print(my_data.tail(n=2))
#columns
my_data.columns
```

```python
#create csv
my_data.to_csv('Test Scores.csv')
```

```python
#read_csv
df = pd.read_csv('Test Scores.csv')
df
#drop columns
df = df.drop(columns = "Unnamed: 0")
df
```
